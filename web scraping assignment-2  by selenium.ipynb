```python
quotes=[]
author=[]
quotes_type=[]
# page scraping
start=0
end=10
for page in range(start,end):
    q_tag=driver.find_elements(By.XPATH,'//a[@class="title"]')   
    for i in q_tag[0:1000]:
        qt=i.text
        quotes.append(qt)
for page in range(start,end):
    a_tag=driver.find_elements(By.XPATH,'//div[@class="author"]') 
    for i in a_tag[0:1000]:
        av=i.text
        author.append(av)
for page in range(start,end):
    qt_tag=driver.find_elements(By.XPATH,'//div[@class="tags"]') 
    for i in qt_tag[0:1000]:
        qt=i.text
        quotes_type.append(qt)

# lenght 
print(len(quotes),len(author),len(quotes_type))
```

```python
# make a dataframe
df8=pd.DataFrame({'QUOTES':quotes,'AUTHOR':author,'QUOTES TYPE':quotes_type})
df8
```
